Primero hay que instalar e importar las librerías:
SymPy para álgebra.
Network para grafos.
Matplotlib para visualizar los grafos.

```
!pip install sympy
!pip install networkx
!pip install matplotlib
import sympy as sp
import networkx as nx
import matplotlib.pyplot as plt
```

Ahora vamos a crear los vectores y matrices en SymPy (`Matrix()`):  
Los vectores se representan como matrices columna o fila.

```
# matriz 2x2
A = sp.Matrix([[1, 2],
               [3, 4]])

# vector columna
v = sp.Matrix([[5],
               [6]])

print("Matriz A:")
sp.pprint(A)

print("\nVector v:")
sp.pprint(v)
```

Vamos a multiplicar vectores con matrices:
Para la multiplicación matriz-vector se usa el operador `*`.

```
resultado = A * v

print("Resultado de A * v:")
sp.pprint(resultado)
```

Potencias de matrices:

Para las potencias de matrices usamos `**`.

Esto es útil en cadenas de Markov para estudiar transiciones en varios pasos.

```
A2 = A**2
A3 = A**3

print("A^2:")
sp.pprint(A2)

print("\nA^3:")
sp.pprint(A3)
```

Solución de sistemas de ecuaciones lineales:

La solución la representaremos como Ax = b

donde:
A es la matriz de coeficientes,
x es el vector de incógnitas,
b es el vector de resultados.

```
# Matriz de coeficientes
A = sp.Matrix([[2, 1],
               [1, 3]])

# Vector de términos independientes
b = sp.Matrix([1, 2])

# Resolver el sistema
x = A.LUsolve(b)

print("Solución del sistema:")
sp.pprint(x)
```

Valores propios (eigenvals) y vectores propios (eigenvects).

```
A = sp.Matrix([[4, 1],
               [2, 3]])

# Valores propios
valores_propios = A.eigenvals()

print("Valores propios:")
print(valores_propios)

# Vectores propios
vectores_propios = A.eigenvects()

print("\nVectores propios:")
print(vectores_propios)
```

Diagramas de transición (pasar de un estdo a otro) con NetworkX:
Nodos (estados), aristas (transiciones).

```
# Crear grafo dirigido
G = nx.DiGraph()

# Agregar estados
G.add_nodes_from(["A", "B", "C"])

# Agregar transiciones con probabilidades
G.add_edge("A", "B", weight=0.5)
G.add_edge("A", "C", weight=0.5)
G.add_edge("B", "A", weight=0.3)
G.add_edge("B", "C", weight=0.7)
G.add_edge("C", "C", weight=1.0)

# Dibujar grafo
pos = nx.spring_layout(G)

nx.draw(G, pos, with_labels=True, node_size=2000, node_color="lightblue")

# Mostrar etiquetas de probabilidad
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

plt.show()
```
